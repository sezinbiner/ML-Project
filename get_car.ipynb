```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
url = "https://www.fordikinciel.com/araba-fiyatlari?p=6"
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')
links = soup.find_all("a", class_="anaResimLink")
links[0]["href"]
for link in links:
  print(link["href"])
  path = "https://www.fordikinciel.com/"
  path = path + link["href"]
  print(path)
  from urllib.request import urlopen
  from bs4 import BeautifulSoup
  url = path
  html = urlopen(url)
  soup = BeautifulSoup(html, 'lxml')
  car = []
  price = soup.find_all('span', class_ = 'price')
  print(price[0].text.split(" ")[-1])
  price = price[0].text.split(" ")[-1]
  car.append(price)
  spans = soup.find_all('span')
  indexes = [17,18,20,21,23,24,25,26,27,28,29,31,32]
  i = 0
  for s in spans:
    if(i in indexes):
      print(s.text.split(" ")[-1])
      car.append(s.text.split(" ")[-1])
    i = i+1
  list_of_cars.append(car)
```
